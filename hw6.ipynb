```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('https://raw.githubusercontent.com/sdami002/datavis/main/complete_severity_of_injury.csv')

# Select a row to create the pie chart for
row_index = 5
row_data = data.iloc[row_index]

# Extract the relevant columns and labels for the pie chart
labels = ['Deaths Male', 'Deaths Female', 'Grievous Injury Male', 'Grievous Injury Female', 'Non Grievous Injury Male', 'Non Grievous Injury Female']
values = [row_data['Deaths Male'], row_data['Deaths Female'], row_data['Grievous Injury Male'], row_data['Grievous Injury Female'], row_data['Non Grievous Injury Male'], row_data['Non Grievous Injury Female']]

# Create the pie chart
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']
plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Injury Severity Distribution for Row {}'.format(row_index))
plt.show()
```
